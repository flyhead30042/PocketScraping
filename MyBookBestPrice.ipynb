## Parse the book grid
1. profle link: <div class="caption"...> <a href=https://share.readmoo.com/book/235506>
2. title: <div class="caption"...> <a title="黑天鵝效應（擴充新版）">

```python
from bs4 import BeautifulSoup
import pandas as pd

soup = BeautifulSoup(res.text, "lxml")
tags = soup.find_all("li", "listGrids-box")
books ={}

blinks=[]
titles=[]
isbns=[]
prices_kobo=[]
for tag in tags:
    for c in tag.contents:
#         print(c)
        try:
            caption = c.find("div", "caption").find("a")
            
            blink = caption["href"]
            print(blink)
            blinks.append(blink)
            title = caption["title"]   
            print(title)    
            titles.append(title)
            
        except Exception as e:
            print("pass")
            pass


books = dict({"title":titles,
             "blink":blinks,
             "isbn":isbns,
             "price_kobo":[],
             "price_google":[]})
```
